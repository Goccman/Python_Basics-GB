**Задача 40**
Работать с файлом california_housing_train.csv, который находится в папке
sample_data. Определить среднюю стоимость дома (median_house_value), где кол-во людей от 0 до 500 (population).


```python
import pandas as pd
```

```python
df = pd.read_csv('sample_data/california_housing_train.csv')
df.head(10)
```

```python
filtered_data = df[(df['population'] >= 0) & (df['population'] <= 500)]
mean_median_house_value = filtered_data['median_house_value'].mean()
```

```python
print("Средняя стоимость дома (median_house_value) при кол-ве людей от 0 до 500:", mean_median_house_value)
```

**Задача 42:**
Узнать какая максимальная households в зоне минимального значения population.

```python
min_population_zone = df.loc[df.population < df.population.quantile(.25)]
max_households = min_population_zone['households'].max()
```

```python
print("Максимальное значение households в зоне с минимальным значением population:", max_households)
```
